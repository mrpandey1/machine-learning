## Introduction To Matplotlib

Its a Plotting library, It allows us to turn our data into pretty visualizations also known as plots or figures

## Why matplotlib

Buit on NumPy arrays (and Python)

Integrates directly with pandas

Can create basic or advanced plots

Simple to use interface (once you get the foundations)

## Matplotlib Example workflow

```python
#import matplotlib
import matplotlib.pyplot as plt

#prepare dataset
x=pd.Series([1,2,3,4])
y=pd.Series([10,20,30,40])

#setup plot
fig,ax=plt.subplots(figsize=(7,7))

#plot data
ax.plot(x,y)

#adding comments and title
ax.set(title='First Plot',xlabel='X-axis',ylabel='Y-axis')

#save the figure
fig.savefig('images/first_pot.png')
```

```python
x=[1,2,3,4,5]
y=[50,60,10,80,77]

fig,ax=plt.subplots(figsize=(7,7))

ax.plot(x,y)

ax.set(title='sample_graph',
      xlabel='X-axis',
       ylabel='Y-axis')

fig.savefig('images/sample_plot2.png')
```

## Histogram

```python
# Histogram
# creating random 1000 values for normal distribution
x = np.random.randn(1000)
fig,ax = plt.subplots()
ax.hist(x);

fig.savefig('images/his1.png')
```

## Two options for subplot

```python
## 1st option
import matplotlib.pyplot as plt
fig,((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,ncols=2,figsize=(10,5))
ax1.plot(x,x/2)
ax2.scatter(np.random.random(10),np.random.random(10))
ax3.bar(shoe_number.keys(),shoe_number.values())
ax4.hist(np.random.randn(1000))
```

```python
# 2nd option
fig,ax=plt.subplots(ncols=2,nrows=2,figsize=(10,5))
ax[0,0].plot(x,x/2)
ax[0,1].scatter(np.random.random(10),np.random.random(10))
ax[1,0].bar(prices.keys(),prices.values())
ax[1,1].hist(np.random.randn(1000))
```

```python
heart_disease=pd.read_csv('Datasets/heart-disease.csv.csv')
```

```python
heart_disease.head()
```

```python
plt.style.use('seaborn')
```

```python
heart_disease.age.plot.hist()
```

```python
heart_disease.plot.hist(subplots=True,figsize=(10,35))
```

```python
over_fifty=heart_disease[heart_disease['age']>50]
```

```python
over_fifty
```

```python
over_fifty.plot(kind='scatter',
               x='age',
               y='chol',
               c='target',
               figsize=(10,6))
```

```python
# oop method
fix,ax = plt.subplots(figsize=(10,6))
over_fifty.plot(kind='scatter',
                x='age',
                y='chol',
                c='target',
                ax=ax);
```

```python
# OOP oriented from scratch
fig,ax=plt.subplots(figsize=(10,6))

# plot data
scatter=ax.scatter(x=over_fifty['age'],
                  y=over_fifty['chol'],
                   c=over_fifty['target']
                  )

#customization
ax.set(title='Heart Disease and cholesterol level',
      xlabel='Age',
       ylabel='Cholesterol'
      )

#add a legend
ax.legend(*scatter.legend_elements(),title='Target')

#add a horizontal line
ax.axhline(over_fifty['chol'].mean(),linestyle='--')
```

```python
over_fifty.head()
```

```python
# subplot
fig,(ax0,ax1)=plt.subplots(nrows=2,ncols=1,figsize=(10,10))

#add data to ax0
scatter=ax0.scatter(x=over_fifty['age'],
                   y=over_fifty['chol'],
                   c=over_fifty['target'])

#customization
ax0.set(title='Heart Disease and cholesterol level',
      xlabel='Age',
       ylabel='Cholesterol'
      )

#add a legend
ax0.legend(*scatter.legend_elements(),title='Target')

ax0.axhline(over_fifty['chol'].mean(),linestyle='--')

#add data to ax1
scatter2=ax1.scatter(x=over_fifty['age'],
                   y=over_fifty['thalach'],
                   c=over_fifty['target'])

#customization
ax1.set(title='Heart Disease and cholesterol level',
      xlabel='Age',
       ylabel='Thalach'
      )

#add a legend
ax1.legend(*scatter.legend_elements(),title='Target')

ax1.axhline(over_fifty['thalach'].mean(),linestyle='--')

# Add a title to the figure
fig.suptitle('Heart Disease Analysis',fontsize=16,fontweight='bold');
```

```python
plt.style.use('seaborn-whitegrid')

# Subplot of chol, age, thalach
fig, (ax0,ax1) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10,10))
# Add data to ax0
scatter = ax0.scatter(x=over_fifty['age'],
                      y=over_fifty['chol'],
                      c=over_fifty['target'],
                      cmap='winter', # this changes color scheme (color map)
                      s=70)  

# customize ax0
ax0.set(title='Heart Disease and Cholesterol Levels',
        xlabel='Age',
        ylabel='Cholestrol')

# Add a legend to ax0
ax0.legend(*scatter.legend_elements(),title='Target')

# Add a mean line
ax0.axhline(over_fifty['chol'].mean(),
            linestyle='--');

# changing the x axis limits
ax0.set_xlim([50,80])

# Add data to ax1
scatter2 = ax1.scatter(x=over_fifty['age'],
                       y=over_fifty['thalach'],
                       c=over_fifty['target'],
                       cmap='winter',
                       s=70)

# Customizing ax1
ax1.set(title='Heart Disease and Heart Levels',
        xlabel='Age',
        ylabel='Thalach')

# Adding legend
ax1.legend(*scatter2.legend_elements(),title='Target');

# Adding mean line
ax1.axhline(over_fifty['thalach'].mean(),linestyle='--');

# changing the x axis limits
ax1.set_xlim([50,80])

# changing the y axis limits
ax1.set_ylim([60,200])

# Add a title to the figure
fig.suptitle('Heart Disease Analysis',fontsize=18,fontweight='bold');
# saving figure
fig.savefig('images/heart-disease-analysis-plot')
```
