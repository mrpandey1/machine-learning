## Introduction To Matplotlib

Its a Plotting library, It allows us to turn our data into pretty visualizations also known as plots or figures

## Why matplotlib

Buit on NumPy arrays (and Python)

Integrates directly with pandas

Can create basic or advanced plots

Simple to use interface (once you get the foundations)

```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.plot()
```

```python
plt.plot([1,2,3,4])#only does on X-axis
```

```python
x=[1,2,3,4]
y=[10,20,30,40]
plt.plot(x,y) # on both X and y axis
```

```python
fig=plt.figure()
ax=fig.add_subplot()
plt.show()
```

```python
fig=plt.figure()
ax=fig.add_axes([1,1,1,1])
ax.plot(x,y)
plt.show()
```

```python
fig,ax=plt.subplots()
ax.plot(x,y)
```

## Matplotlib Example workflow

```python
#import matplotlib
import matplotlib.pyplot as plt

#prepare dataset
x=pd.Series([1,2,3,4])
y=pd.Series([10,20,30,40])

#setup plot
fig,ax=plt.subplots(figsize=(7,7))

#plot data
ax.plot(x,y)

#adding comments and title
ax.set(title='First Plot',xlabel='X-axis',ylabel='Y-axis')

#save the figure
fig.savefig('images/first_pot.png')
```

```python
x=[1,2,3,4,5]
y=[50,60,10,80,77]

fig,ax=plt.subplots(figsize=(7,7))

ax.plot(x,y)

ax.set(title='sample_graph',
      xlabel='X-axis',
       ylabel='Y-axis')

fig.savefig('images/sample_plot2.png')
```

## Making figures with NumPy arrays

- Line Plot
- Scatter Plot
- Bar Plot
- Histogram
- Sublots

## Line Plot

```python
x=np.linspace(0,10,100)

fig,ax=plt.subplots(figsize=(7,7))

ax.plot(x,x**2)

fig.savefig('images/Lineplot.png')
```

## Scatter Plot

```python
fig,ax=plt.subplots()
ax.scatter(x,np.exp(x))

fig.savefig('images/Scatterplot.png')
```

```python
fig,ax=plt.subplots(figsize=(7,7))
ax.scatter(x,np.sin(x))

fig.savefig('images/Scatterplot2.png')
```

## Bar Plot

```python
prices={'mi a3':13000,'realme 6 pro':15000,'Moto e4 plus':10000}
fig,ax=plt.subplots(figsize=(7,7))
ax.bar(prices.keys(),prices.values())
```

```python
shoe_number={'Nishant':11,'Prashant':10,'Ganesh':7,'Radha':6}
fig,ax=plt.subplots()
ax.bar(shoe_number.keys(),shoe_number.values())
```

```python
## horizontal bars
fig,ax=plt.subplots()
ax.barh(list(shoe_number.keys()),list(shoe_number.values()))
```

## Histogram

```python
x=np.random.randn(1000)
fig,ax=plt.subplots()
ax.hist(x)

fig.savefig('images/his1.png')
```
