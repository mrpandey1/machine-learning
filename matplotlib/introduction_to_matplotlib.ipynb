## Introduction To Matplotlib

Its a Plotting library, It allows us to turn our data into pretty visualizations also known as plots or figures

## Why matplotlib

Buit on NumPy arrays (and Python)

Integrates directly with pandas

Can create basic or advanced plots

Simple to use interface (once you get the foundations)

```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4]
y=[10,20,30,40]
plt.plot(x,y) # on both X and y axis
```

```python
fig=plt.figure()
ax=fig.add_axes([1,1,1,1])
ax.plot(x,y)
plt.show()
```

## Matplotlib Example workflow

```python
#import matplotlib
import matplotlib.pyplot as plt

#prepare dataset
x=pd.Series([1,2,3,4])
y=pd.Series([10,20,30,40])

#setup plot
fig,ax=plt.subplots(figsize=(7,7))

#plot data
ax.plot(x,y)

#adding comments and title
ax.set(title='First Plot',xlabel='X-axis',ylabel='Y-axis')

#save the figure
fig.savefig('images/first_pot.png')
```

```python
x=[1,2,3,4,5]
y=[50,60,10,80,77]

fig,ax=plt.subplots(figsize=(7,7))

ax.plot(x,y)

ax.set(title='sample_graph',
      xlabel='X-axis',
       ylabel='Y-axis')

fig.savefig('images/sample_plot2.png')
```

## Line Plot

```python
x=np.linspace(0,10,100)

fig,ax=plt.subplots(figsize=(7,7))

ax.plot(x,x**2)

fig.savefig('images/Lineplot.png')
```

## Scatter Plot

```python
fig,ax=plt.subplots()
ax.scatter(x,np.exp(x))

fig.savefig('images/Scatterplot.png')
```

```python
fig,ax=plt.subplots(figsize=(7,7))
ax.scatter(x,np.sin(x))

fig.savefig('images/Scatterplot2.png')
```

## Bar Plot

```python
prices={'mi a3':13000,'realme 6 pro':15000,'Moto e4 plus':10000}
fig,ax=plt.subplots(figsize=(7,7))
ax.bar(prices.keys(),prices.values(), color=(0.2, 0.4, 0.6, 0.6))
```

```python
shoe_number={'Nishant':11,'Prashant':10,'Ganesh':7,'Radha':6}
fig,ax=plt.subplots()
ax.bar(shoe_number.keys(),shoe_number.values())
```

```python
## horizontal bars
fig,ax=plt.subplots()
ax.barh(list(shoe_number.keys()),list(shoe_number.values()))
```

## Histogram

```python
# Histogram
# creating random 1000 values for normal distribution
x = np.random.randn(1000)
fig,ax = plt.subplots()
ax.hist(x);

fig.savefig('images/his1.png')
```

## Two options for subplot

```python
## 1st option
import matplotlib.pyplot as plt
fig,((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,ncols=2,figsize=(10,5))
ax1.plot(x,x/2)
ax2.scatter(np.random.random(10),np.random.random(10))
ax3.bar(shoe_number.keys(),shoe_number.values())
ax4.hist(np.random.randn(1000))
```

```python
# 2nd option
fig,ax=plt.subplots(ncols=2,nrows=2,figsize=(10,5))
ax[0,0].plot(x,x/2)
ax[0,1].scatter(np.random.random(10),np.random.random(10))
ax[1,0].bar(prices.keys(),prices.values())
ax[1,1].hist(np.random.randn(1000))
```

```python
car_sales.head()
```

```python
car_sales['Price'] = car_sales['Price'].str.replace('[\$\,\.]','')
car_sales['Price']=car_sales['Price'].str[:-2]
```

```python
car_sales
```

```python
## adding sales date column
car_sales['Sales Date']=pd.date_range('1/1/2020',periods=
                                    len(car_sales))
car_sales
```

```python
car_sales['Total sum']=car_sales['Price'].astype(int).cumsum()
car_sales.head()
```

```python
type(car_sales['Total sum'][0])
```

```python
car_sales.plot(x='Sales Date',y='Total sum');
```

```python
car_sales['Price']=car_sales['Price'].astype(int)
car_sales.plot(x='Odometer (KM)',y='Price',kind='scatter')
```

```python
car_sales.plot(x='Make',y='Price',kind='bar')
```
